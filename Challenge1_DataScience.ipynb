# **Importación de datos**



## **Empezaremos realizando al importación de librerias PANDAS y Matplotlib.pyplot**

```python
# Importación de la libreria PANDAS
import pandas as pd
import matplotlib.pyplot as plt

# URL de los archivos CSV de cada tienda
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Creacion de un dataframe por cada tienda a partir de la url de cada csv
tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Muestra de las primeras 5 filas del dataframe Tienda1
tienda1.head()
```

# **1. Análisis de facturación**



```python
# Cálculo de la ventas por tiendas
df_tiendas_suma = df_tiendas.groupby('Tienda')['Precio'].sum()
df_tiendas_suma
```

```python
# figure()--> crea un anueva figura/grafico con dimensiones de 10' de ancho y 6' de alto
plt.figure(figsize=(10,6))
# bar()--> crea gráfico de barras usando los indices del dataframe como etiquetas del eje X,
# y los valores como altura,y asigna colores especificos a cada barra
plt.bar(df_tiendas_suma.index,df_tiendas_suma.values,color=['#639CF8', '#BBD4FC','#B9B8FF', '#8FB8FA'])
# title()--> Establece el título del grafico con tamaño de fuente 15
plt.title('Ventas Totales por Tienda',fontsize=15,fontweight='bold')
# xlabel()--> define la etiqueta del eje X con tamaño de fuente 15
plt.xlabel('Tiendas',fontsize=15)
# ylabel()--> define la etiqueta del eje y con tamaño de fuente 15
plt.ylabel('Ventas',fontsize=15)
# grid(): Añade líneas de cuadrícula solo en el eje Y, con estilo de línea discontinua (--)
# y transparencia del 70% (alpha=0.7).
plt.grid(axis='y',linestyle='--',alpha=0.7)
'''
enumerate() y text(): Itera sobre los valores y añade etiquetas de texto encima de cada barra mostrando
el valor en formato monetario ($) con separadores de miles, sin decimales, centrado horizontalmente y
alineado en la parte inferior.
'''
for index,value in enumerate(df_tiendas_suma.values):
    plt.text(index, value, f'${value:,.0f}', ha='center', va='bottom', fontsize=9,fontweight='bold')

# ticklabel_format()-->Establece el formato de las etiquetas del eje Y en estilo 'plain' (sin notación científica)
plt.ticklabel_format(style='plain', axis='y')
# tight_layout()-->Ajusta automáticamente los elementos del gráfico para evitar superposiciones y optimizar el espacio.
plt.tight_layout()
# show()--> Muestra el gráfico en pantalla.
plt.show()
```

# **2. Ventas por categoría**

Para evaluar las ventas por categoría,procederemos a crear un dataframe donde esten agrupados por Tienda y Categoría mostrando las ventas totales y sus respectivas cantidades .

```python

df_categoria_ventas = df_tiendas.groupby(['Tienda','Categoría'])['Precio'].sum()
```

```python
df_categoria_numventas = df_tiendas.groupby(['Tienda','Categoría'])['Precio'].count()
```

```python
df_vpc = pd.concat([df_categoria_ventas.rename('Ventas'),df_categoria_numventas.rename('Cant.Ventas')], axis=True)
df_vpc
```

```python
df_categoria_numventas
```

```python
import matplotlib.pyplot as plt

# Define a color palette for the pie chart slices
pie_colors = plt.cm.Paired.colors

# Iterate through each unique store in df_vpc
for store_name in df_vpc.index.get_level_values('Tienda').unique():
    # Get the 'Cant.Ventas' for each category in the current store
    store_category_sales_count = df_vpc.loc[store_name]['Cant.Ventas']
    total_sales_count = store_category_sales_count.sum()

    # Create the pie chart using subplots to get an axes object for more control
    fig, ax = plt.subplots(figsize=(9, 5))
    wedges, texts, autotexts = ax.pie(
        store_category_sales_count.values,
        autopct='%1.1f%%', # Initial percentage format
        startangle=90,
        colors=pie_colors,
        pctdistance=0.85 # Distance of percentage labels from center
    )

    # Customize autotexts to include both percentage and value
    for i, autotext in enumerate(autotexts):
        value = store_category_sales_count.values[i]
        percentage = (value / total_sales_count) * 100 # Calculate percentage directly
        autotext.set_text(f'{percentage:.1f}%\n({value:,.0f})') # Format: Percentage (Value)
        autotext.set_color('black')
        autotext.set_fontsize(9)
        autotext.set_fontweight('bold')

    ax.set_title(f'Distribución de Cantidad de Ventas por Categoría en {store_name}', fontsize=12, fontweight='bold')
    ax.axis('equal') # Ensures the pie chart is circular.

    # Add legend to the right of the chart
    ax.legend(labels=store_category_sales_count.index, loc='center left', bbox_to_anchor=(1, 0.5))

    plt.tight_layout()
    plt.show()
```

# **3. Calificación promedio de la tienda**


```python
df_cpt = df_tiendas.groupby('Tienda',as_index=False)['Calificación'].agg(Promedio='mean', Mínimo='min', Máximo='max', Total_Registros='count')
df_cpt = df_cpt.sort_values(by='Promedio',ascending=False)
df_cpt
```

```python
plt.figure(figsize=(10,7))
plt.bar(df_cpt['Tienda'],df_cpt['Promedio'],
        color=['#639CF8', '#BBD4FC','#B9B8FF', 'red'])
plt.title('Promedio por Tienda',fontsize=15,fontweight='bold')
plt.xlabel('Tiendas',fontsize=15)
plt.ylabel('Promedio',fontsize=15)
plt.grid(axis='y',linestyle='--',alpha=0.7)

for index,value in enumerate(df_cpt.values):
    plt.text(index, value[1], f'{value[1]:.2f}', ha='center', va='bottom', fontsize=9,fontweight='bold')
plt.tight_layout()
plt.show()
```

## **Análisis de calificación por tienda:**
Luego de realizar el tratamiento de los datos mediante el uso del metodo groupby,creamos un dataframe en el cual se observa la calificación promedio obtenida por cada tienda ,adicional a ello se resalto de color rojo la tienda con menor calificación,de ello observamos que la **" Tienda 1 "** es la que presenta menor calificación con un valor de **3.98**.
-  La **tienda 3** presenta el puntaje mas alto de calificación **4.05**

# **4. Productos más y menos vendidos**

```python
df_productos_por_tienda_categoria = df_tiendas.groupby(['Tienda', 'Producto', 'Categoría']).size().reset_index(name='Cantidad_Vendida')

for store in df_productos_por_tienda_categoria['Tienda'].unique():
    print(f"\n--------- Productos, Categoría y Cantidad Vendida para {store} -----------")

    # Filtrar por tienda y mostrar la tabla completa
    productos_en_tienda = df_productos_por_tienda_categoria[df_productos_por_tienda_categoria['Tienda'] == store]

    # Ordenar por Cantidad_Vendida para una mejor visualización
    productos_en_tienda_sorted = productos_en_tienda.sort_values(by='Cantidad_Vendida', ascending=False).reset_index(drop=True)

    # Mostrar los 3 productos con más cantidad vendida
    print(f"\n--- ✅ Los 3 productos con más cantidad vendida en {store} ---")
    display(productos_en_tienda_sorted.head(3))

    # Mostrar los 3 productos con menos cantidad vendida
    print(f"\n--- ⬇️ Los 3 productos con menos cantidad vendida en {store} ---")
    display(productos_en_tienda_sorted.tail(3))
```
